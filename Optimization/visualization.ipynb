```python
def plot_planning(planning):
    m_count = 0
    plt.figure()
    for machine in planning.keys():
        m_count += 1
        x = []
        y = []
        for time in planning[machine]:
            x.append(time)
            y.append(machine)
        
        plt.scatter(x,y)

for i in range(5):
    plot_planning(plannings[i])
```
